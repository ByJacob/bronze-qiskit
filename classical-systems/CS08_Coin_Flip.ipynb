### Moneta uczciwa

Moneta ma dwie strony: <i>Orzeł</i> i <i>Reszka</i>.

Po rzuceniu monetą otrzymujemy albo Orła, albo Reszkę. Możemy reprezentować te dwa różne przypadki za pomocą pojedynczego bitu:
- 0 reprezentuje Orła
- 1 reprezentuje Reszkę


### Rzucanie uczciwą monetą

Jeśli nasza moneta jest uczciwa, to prawdopodobieństwa wypadnięcia Orła i Reszki są równe:

$ p= \dfrac{1}{2} = 0.5 $.

Rzucanie uczciwą monetą można przedstawić jako operator:
- $ FairCoin(Heads) = \frac{1}{2} Heads + \frac{1}{2}Tails $
- $ FairCoin(Tails) \mspace{10mu} = \frac{1}{2} Heads + \frac{1}{2}Tails $

Oto jego reprezentacja tabelaryczna:

$
FairCoin = \begin{array}{c|cc} \hookleftarrow & \mathbf{Heads} & \mathbf{Tails} \\ \hline \mathbf{Heads} & \dfrac{1}{2} & \dfrac{1}{2} \\  \mathbf{Tails} & \dfrac{1}{2} & \dfrac{1}{2}  \end{array} 
$

Oto ta sama tabela, używając 0 i 1 jako stanów:

$
FairCoin = \begin{array}{c|cc} \hookleftarrow & \mathbf{0} & \mathbf{1} \\ \hline \mathbf{0} & \dfrac{1}{2} & \dfrac{1}{2} \\  \mathbf{1} & \dfrac{1}{2} & \dfrac{1}{2}  \end{array} 
$

### Rzucanie urojonym monetą

Nasza moneta może być urojona.

Na przykład prawdopodobieństwo wypadnięcia orła jest większe niż prawdopodobieństwo wypadnięcia reszki.

Oto przykład:

$
BiasedCoin = \begin{array}{c|cc} \hookleftarrow & \mathbf{Heads} & \mathbf{Tails} \\ \hline \mathbf{Heads} & 0.6 & 0.6 \\  \mathbf{Tails} & 0.4 & 0.4  \end{array}
$

Korzystając z 0 i 1 jako stanów:

$
BiasedCoin = \begin{array}{c|cc} \hookleftarrow & \mathbf{0} & \mathbf{1} \\ \hline \mathbf{0} & 0.6 & 0.6\\  \mathbf{1} & 0.4 & 0.4 \end{array}
$

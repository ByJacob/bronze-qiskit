<a id="task1"></a>
<h3> Zadanie 1 </h3>

Oblicz stan systemu złożonego.


<h3>Rozwiązanie</h3>

Prawdopodobieństwa stanów Asji i Balvisa wynoszą odpowiednio $ \myvector{\frac{1}{2} \\ \frac{1}{2}} $ i $ \myvector{1 \\ 0} $.

Stan układu złożonego to $ \myvector{\frac{1}{2} \\ \frac{1}{2}} \otimes \myvector{1 \\ 0} = \myvector{\frac{1}{2} \myvector{1 \\ 0} \\ \frac{1}{2} \myvector{1 \\ 0} } = \myvector{\frac{1}{2} \\ 0 \\ \frac{1}{2} \\ 0 } = \frac{1}{2} \pstate{00} + \frac{1}{2} \pstate{10}  $.


<a id="task2"></a>
<h3> Zadanie 2 </h3>

Sprawdź, czy stan systemu złożonego to $ \frac{1}{2} \pstate{00} + \frac{1}{2} \pstate{11} $ po operacji CNOT.


### Rozwiązanie

$ \text{CNOT} \big( \frac{1}{2} \ket{00} + \frac{1}{2} \ket{10} \big) = \text{CNOT}  \frac{1}{2} \ket{00} + \text{CNOT} \frac{1}{2} \ket{10} = \frac{1}{2} \ket{00} + \frac{1}{2} \ket{11} $.

Alternatywnie,

$ \text{CNOT}  \begin{pmatrix} \frac{1}{2} \\ 0 \\ \frac{1}{2} \\ 0 \end{pmatrix} =  \begin{pmatrix} \frac{1}{2} \\ 0 \\ 0 \\ \frac{1}{2} \end{pmatrix} = \frac{1}{2} \ket{00} + \frac{1}{2} \ket{11} $.


<a id="task4"></a>
<h3> Zadanie 4 </h3>

Pokaż, że nie jest możliwe przedstawienie wektora $ \myvector{ \frac{1}{2} \\ 0 \\ 0 \\ \frac{1}{2} } $ jako $ \myvector{a\\b} \otimes \myvector{c\\d} $ dla pewnych liczb $ a, b, c $ i $ d $.

<i>Zauważ, że korelacja w ogólności może nie być łatwo pokazana jak w tym podstawowym przykładzie. </i>


### Rozwiązanie

$ \myvector{a\\b} \otimes \myvector{c\\d} = \myvector{ a  \myvector{c\\d}  \\ b  \myvector{c\\d}  } = \myvector{ ac \\ ad \\ bc \\ bd } $. 
 
Ponieważ $ ac = \frac{1}{2} $, możemy wywnioskować, że $a$ i $c$ są liczbami różnymi od zera.

Ponieważ $ bd = \frac{1}{2} $, możemy wywnioskować, że $b$ i $d$ są liczbami różnymi od zera.

Zatem $ ad $ i $ bc $ są również liczbami różnymi od zera. 

Jednakże jest to sprzeczność: $ \myvector{ ac \\ ad \\ bc \\ bd } = \myvector{\frac{1}{2} \\ 0 \\ 0 \\ \frac{1}{2}} $, więc $ ad $ i $ bc $ muszą być zerami.


<a id="task7"></a>
<h3> Zadanie 7 </h3>

Oblicz nowy stan złożonego systemu zdefiniowanego na trzech bitach po operatorze CNOT zdefiniowanym w Zadaniu 5 lub Zadaniu 6.

$ \frac{1}{2} \pstate{000} + \frac{1}{2} \pstate{110} \rightarrow ? $


<h3>Rozwiązanie</h3>


$ \frac{1}{2} \pstate{000} + \frac{1}{2} \pstate{\mathbf{\color{red}{1}}1\mathbf{\color{blue}{0}}} \rightarrow \frac{1}{2} \pstate{000} + \frac{1}{2} \pstate{\mathbf{\color{red}{1}}1\mathbf{\color{blue}{1}}} $.


<a id="task8"></a>
<h3> Zadanie 8 </h3>

Załóżmy, że Daria miała 10 centów w stanie 1, a jej moneta została również połączona z układem złożonym.

Jaki jest stan nowego układu złożonego?

Następnie operator CNOT jest zastosowany do monety Curie i monety Darii, gdzie moneta Curie jest bitem kontrolnym.

Jaki jest nowy stan układu złożonego po zastosowaniu tego operatora CNOT?

<i>Zauważ, że korelacja z nowym bitem może być stworzona poprzez zastosowanie bramki CNOT pomiędzy dowolnym bitem już znajdującym się w korelacji oraz nowym bitem, gdzie nowy bit jest bitem docelowym.</i>


<h3>Rozwiązanie</h3>

Stan nowego systemu złożonego to

$ \mypar{ \frac{1}{2} \pstate{000} + \frac{1}{2} \pstate{111} } \otimes \pstate{1} = \frac{1}{2} \pstate{0001} + \frac{1}{2} \pstate{1111}  $.


Po zastosowaniu bramki CNOT zdefiniowanej na trzecim bicie (sterującym) i czwartym bicie (docelowym):

$ \frac{1}{2} \pstate{0001} + \frac{1}{2} \pstate{11\mathbf{\color{red}{1}}\mathbf{\color{blue}{1}}} \rightarrow \frac{1}{2} \pstate{0001} + \frac{1}{2}\pstate{11\mathbf{\color{red}{1}}\mathbf{\color{blue}{0}}} $.

Teraz wszystkie cztery bity są skorelowane. Patrząc na stan dowolnego bitu, możemy poznać stany pozostałych bitów.

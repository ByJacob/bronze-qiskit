Załóżmy, że mamy monetę ustawioną na orła z prawdopodobieństwem $ 0.3 $, która jest rzucana raz, ale nie widzimy wyniku.

Mimo że wyląduje ona na orle lub reszce, nasza informacja na temat wyniku jest probabilistyczna: Jest to orzeł z prawdopodobieństwem $ 0.3 $ i reszka z prawdopodobieństwem $ 0.7 $.

Jeśli potraktujemy tę monetę jako system o dwóch stanach (orzeł i reszka: odpowiednio, stany 0 i 1), to możemy powiedzieć, że po jednej iteracji jest w stanach 0 i 1 z prawdopodobieństwem 0.3 i 0.7, odpowiednio.

Ogólnie rzecz biorąc, bit probabilistyczny jest w stanach 0 i 1 z prawdopodobieństwem $ p $ i $ 1-p $, gdzie $ p $ jest liczbą między 0 a 1. Zauważ, że jeśli $ p=1 $ lub $ p=0 $, to bit staje się deterministyczny.

Jeśli powyższa ustawiona moneta zostanie rzucona jeszcze raz, prawdopodobieństwa otrzymania orła i reszki (lub bycia w stanach 0 i 1) pozostaną takie same. W następnej grze, korzystając z dwóch ustawionych monet, będziemy mogli uzyskać różne prawdopodobieństwa dla stanów 0 i 1. Nasz przyjaciel Asja ma jeden euro i jednego centa.

Obydwie monety są stronnicze, a prawdopodobieństwa wypadnięcia orła i reszki są następujące:
<ul>
    <li> jeden euro: orzeł z prawdopodobieństwem $ 0.6 $ i reszka z prawdopodobieństwem $ 0.4 $. </li>
    <li> jeden cent: orzeł z prawdopodobieństwem $ 0.3 $ i reszka z prawdopodobieństwem $ 0.7 $. </li>
</ul>

Asja rzuca swoimi monetami zgodnie z następującym <b>protokołem</b>:
<ol> 
    <li> zaczyna od rzucenia jednym euro[*]; </li>
    <li> za każdym razem, gdy wypadnie orzeł, w następnej rundzie rzuca jednym euro; oraz, </li>
    <li> za każdym razem, gdy wypadnie reszka, w następnej rundzie rzuca jednym centem. </li>
</ol>

Korzystając z jednego bita, podsumowujemy wszystkie przejścia tej gry w następujący sposób:

$
GameCoins = \begin{array}{c|cc} \hookleftarrow & \mathbf{Orzeł} & \mathbf{Reszka} \\ \hline \mathbf{Orzeł} & 0.6 & 0.3\\  \mathbf{Reszka} & 0.4 & 0.7  \end{array} = \begin{array}{c|cc} \hookleftarrow & \mathbf{0} & \mathbf{1} \\ \hline \mathbf{0} & 0.6 & 0.3 \\  \mathbf{1} & 0.4 & 0.7  \end{array}
$

[*] Powinniśmy ustalić warunek początkowy. W przeciwnym razie Asja nie może wybrać żadnej z monet na początku gry.

````python
#
# OUR SOLUTION
#

# initial condition:
# Asja will start with one euro,
#    and so, we assume that the probability of having head is 1 at the beginning.
prob_head = 1
prob_tail = 0


#
# first coin-flip
#

# the new probability of head is calculated by using the first row of table
new_prob_head = prob_head * 0.6 + prob_tail * 0.3

# the new probability of tail is calculated by using the second row of the table
new_prob_tail = prob_head * 0.4 + prob_tail * 0.7

# update the probabilities for the second round
prob_head = new_prob_head
prob_tail = new_prob_tail

#
# second coin-flip
#
# we do the same calculations

new_prob_head = prob_head * 0.6 + prob_tail * 0.3
new_prob_tail = prob_head * 0.4 + prob_tail * 0.7

prob_head = new_prob_head
prob_tail = new_prob_tail

#
# third coin-flip
#
# we do the same calculations

new_prob_head = prob_head * 0.6 + prob_tail * 0.3
new_prob_tail = prob_head * 0.4 + prob_tail * 0.7

prob_head = new_prob_head
prob_tail = new_prob_tail

# print prob_head and prob_tail
print("the probability of getting head after 3 coin tosses is",prob_head)
print("the probability of getting tail after 3 coin tosses is",prob_tail)
````

Rozpoczynamy od wyjaśnienia (autorstwa Daniela Reitznera).

Nasz ojciec przygotowuje lunch dla mojej siostry i dla mnie. Umieszcza w naszych pojemnikach albo sałatkę z makaronem, albo sałatkę kuskus.

Do momentu otwarcia pojemników nie wiemy, co mamy na lunch. Jednakże, gdy otworzę swój pojemnik, poznam również lunch mojej siostry, i vice versa.

To jest <i>korelacja</i>.


Teraz przejdziemy do bardziej formalnego wyjaśnienia.

Asja ma jedno euro, a Balvis ma jeden cent.

Obie monety są uczciwe i nie widzimy ich monet.

Zaczynają od "orła" (stan 0): $ \vzero $ i $ \vzero $, odpowiednio.
<br>
Asja rzuca swoją monetą raz. Następnie prawdopodobne stany Asji i Balvisa są odpowiednio $ \myvector{\frac{1}{2} \\ \frac{1}{2}} $ i $ \myvector{1 \\ 0} $.

Wykonują operator Controlled-NOT na swoich monetach, który działa na dwóch bitach:

$$
    CNOT = \CNOT.
$$

<i> Efektem działania CNOT jest to, że jeśli stan pierwszego bitu wynosi 1, to wartość drugiego bitu zostaje odwrócona.

Tutaj pierwszy bit jest bitem kontrolnym, a drugi jest bitem docelowym.</i>

Korzystając z stanów bazowych:
<ul style="font-size:16px;">
    <li>$CNOT \pstate{00} = \pstate{00}$</li>
    <li>$CNOT \pstate{01} = \pstate{01}$</li>
    <li>$CNOT \pstate{\mathbf{\color{red}{1}\color{blue}{0}}} = \pstate{\mathbf{\color{red}{1}\color{blue}{1}}} $</li>
    <li>$CNOT \pstate{\mathbf{\color{red}{1}\color{blue}{1}}} = \pstate{\mathbf{\color{red}{1}\color{blue}{0}}} $</li>
</ul>

<h3> Korelacja z więcej niż dwoma bitami</h3>

Curie ma monetę o wartości 50 centów w stanie 0.

Wtedy, stan systemu złożonego zdefiniowanego przez te trzy monety będzie równy

$ \mypar{ \frac{1}{2} \pstate{00} + \frac{1}{2} \pstate{11}  } \otimes \pstate{0} = 
\frac{1}{2} \pstate{000} + \frac{1}{2} \pstate{110} $.

Co się stanie, jeśli operator CNOT zostanie zastosowany do monety Asji i monety Curie, gdzie moneta Asji jest kontrolną, a moneta Curie jest docelową?

CNOT to operator zdefiniowany dla dwóch bitów. Dlatego może być zastosowany do dowolnej pary bitów.

W takim przypadku wartość dowolnego innego bitu w systemie złożonym nie zmienia się.

Można zadać pytanie, jaki jest jego zapis macierzowy zdefiniowany na trzy bity.

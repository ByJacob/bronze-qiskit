### Reprezentacja wektorowa

Załóżmy, że mamy system z 4 rozróżnialnymi stanami: $ s_1 $, $s_2 $, $s_3$ i $s_4$.

Oczekujemy, że system będzie w dowolnym momencie w jednym z nich.

Rozmawiając o prawdopodobieństwach, mówimy, że system jest w jednym ze stanów z prawdopodobieństwem 1, a w każdym innym stanie z prawdopodobieństwem 0.

Korzystając z naszej reprezentacji kolumnowej, możemy pokazać każdy stan jako wektor kolumnowy (korzystając z wektorów w bazie standardowej $ \mathbb{R}^4 $):

$
    e_1 = \begin{bmatrix} 1\\ 0 \\ 0 \\ 0 \end{bmatrix}, e_2 = \begin{bmatrix} 0 \\ 1 \\ 0 \\ 0 \end{bmatrix}, e_3 = \begin{bmatrix} 0 \\ 0 \\ 1 \\ 0 \end{bmatrix}, 
    \text{ oraz } e_4 = \begin{bmatrix} 0 \\ 0 \\ 0 \\ 1 \end{bmatrix}.
$


Ta reprezentacja pomaga nam przedstawić informacje o systemie, gdy znajduje się on w więcej niż jednym stanie z określonymi prawdopodobieństwami.

Przypomnijmy sobie sytuację, gdy monety są rzucane w tajemnicy.

Na przykład, załóżmy, że system znajduje się w stanach $ s_1 $, $ s_2 $, $ s_3 $ i $ s_4 $ z prawdopodobieństwami odpowiednio $ 0.20 $, $ 0.25 $, $ 0.40 $ i $ 0.15 $.

(<i>Całkowite prawdopodobieństwo musi wynosić 1, czyli $ 0.20+0.25+0.40+0.15 = 1.00 $</i>)

Wówczas możemy powiedzieć, że system znajduje się w następującym stanie probabilistycznym:

$ 0.20 \cdot e_1 + 0.25 \cdot e_2 + 0.40 \cdot e_3 + 0.15 \cdot e_4 $

$ =  0.20 \cdot \myvector{1\\ 0 \\ 0 \\ 0} + 0.25 \cdot \myvector{0\\ 1 \\ 0 \\ 0} + 0.40 \cdot \myvector{0\\ 0 \\ 1 \\ 0} + 0.15 \cdot  \myvector{0\\ 0 \\ 0 \\ 1}  $

$ = \myvector{0.20\\ 0 \\ 0 \\ 0} +   \myvector{0\\ 0.25 \\ 0 \\ 0} +  \myvector{0\\ 0 \\0.40 \\ 0} +  \myvector{0\\ 0 \\ 0 \\ 0.15 } = \myvector{ 0.20 \\ 0.25 \\ 0.40 \\ 0.15 }, $

gdzie suma elementów musi wynosić 1.

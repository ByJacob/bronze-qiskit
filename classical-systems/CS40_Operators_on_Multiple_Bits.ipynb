### Operatory jednobitowe


Kiedy mamy dwa bity, to nasz system ma cztery stany, a każdy operator systemu może być zdefiniowany jako macierz o wymiarach $(4 \times 4)$.

Na przykład, jeśli zastosujemy operator probabilistyczny $ M =  \mymatrix{c}{ 0.3 & 0.6 \\ 0.7 & 0.4  } $ na drugim bicie, to jak możemy reprezentować odpowiadającą mu macierz o wymiarach $ (4 \times 4) $?

Ponieważ chcemy szczególnie, aby operator $ M $ działał na drugim bicie i jednocześnie pozostawił pierwszy bit niezmieniony, piszemy macierz $4 \times 4$ jako iloczyn tensorowy operatora identycznościowego oraz macierzy M, tak aby

$ I \otimes M = \I \otimes  \mymatrix{c}{ 0.3 & 0.6 \\ 0.7 & 0.4  } = \mymatrix{cccc} { 0.3 & 0.6 & 0 & 0 \\ 0.7 & 0.4 & 0 & 0 \\ 0 & 0 & 0.3 & 0.6 \\ 0 & 0& 0.7 & 0.4 }.   $

W ten sposób operator identycznościowy i operator M działają odpowiednio na pierwszym i drugim bicie, ze względu na definicję iloczynu tensorowego.

### Operatory dwubitowe


Zaczynamy od prostego przykładu.

Mamy trzy bity i stosujemy operator probabilistyczny 
$ M = \mymatrix{rrrr}{0.05 & 0 & 0.70 & 0.60 \\ 0.45 & 0.50 & 0.20 & 0.25 \\ 0.20 & 0.35 & 0.10 & 0 \\ 0.30 & 0.15 & 0 & 0.15 } $ do pierwszego i drugiego bitu. Następnie odpowiednia macierz o wymiarach $ (8 \times 8) $ to $ M \otimes I $, gdzie $I$ oznacza dwuwymiarową macierz jednostkową $(2 \times 2)$.

Jeśli $ M $ jest stosowane do drugiego i trzeciego bitu, to odpowiadająca macierz to $ I \otimes M $.


**Co jeśli $ M $ zostanie zastosowane do pierwszego i trzeciego bitu?**

Wybieramy przykładową przejście: w $ M $ dane jest, że $ \greenbit{0} \brownbit{1}  \xrightarrow{0.35} \greenbit{1} \brownbit{0}  $.  
- To znaczy, że gdy pierwszy bit jest równy 0, a trzeci bit jest równy 1, pierwszy bit jest ustawiany na 1, a trzeci bit jest ustawiany na 0 z prawdopodobieństwem 0.35:

$$ \myarray{ccccc}{\mbox{pierwszy-bit} & \mbox{trzeci-bit} & prawdopodobieństwo & \mbox{pierwszy-bit} & \mbox{trzeci-bit} \\ \greenbit{0} & \brownbit{1} & \xrightarrow{0.35} & \greenbit{1} & \brownbit{0} }  $$

- Wprowadzamy drugi bit na obrazek, zakładając że jest on operatorem identycznościowym:

$$ 
\myarray{ccccccc}{
\mbox{pierwszy-bit} & \mbox{drugi-bit} & \mbox{trzeci-bit} & prawdopodobieństwo & \mbox{pierwszy-bit} & \mbox{drugi-bit} & \mbox{trzeci-bit} \\ 
\greenbit{0} & \bluebit{0} & \brownbit{1} & \xrightarrow{0.35} & \greenbit{1} & \bluebit{0} & \brownbit{0}  \\
\greenbit{0} & \bluebit{1} & \brownbit{1} & \xrightarrow{0.35} & \greenbit{1} & \bluebit{1} & \brownbit{0} \\
\\ \hline \\
\greenbit{0} & \bluebit{0} & \brownbit{1} & \xrightarrow{0} & \greenbit{1} & \bluebit{1} & \brownbit{0}  \\
\greenbit{0} & \bluebit{1} & \brownbit{1} & \xrightarrow{0} & \greenbit{1} & \bluebit{0} & \brownbit{0}
}
$$

### Operatory kontrolowane


Forma macierzowa operatora CNOT (Controlled-NOT) jest następująca:

$ CNOT = \mymatrix{cc|cc}{ 1 & 0 & 0 & 0 \\ 0 & 1 & 0 & 0 \\ \hline 0 & 0 & 0 & 1 \\ 0 & 0 & 1 & 0 } 
= \mymatrix{c|c}{ I & \mathbf{0} \\ \hline \mathbf{0} & X},
$
gdzie $ X $ oznacza operator NOT.

Podobnie, dla danego jednobitowego operatora $ M $, możemy zdefiniować operator kontrolowany-$M$ (gdzie pierwszy bit jest bitem kontrolnym, a drugi bit jest bitem docelowym) w następujący sposób:

$ CM = \mymatrix{c|c}{ I & \mathbf{0} \\ \hline \mathbf{0} & M } $

Z definicji:
* gdy pierwszy bit jest równy 0, to na drugim bicie stosowana jest identyczność, a
* gdy pierwszy bit jest równy 1, to na drugim bicie stosowany jest operator $ M $. 

Obserwujemy, że macierz $ CM $ ma ładną formę, ponieważ pierwszy bit jest bitem kontrolnym. Macierz $ CM $ podana powyżej jest podzielona na cztery podmacierze na podstawie stanów pierwszego bita. Wtedy możemy zauważyć, że
* wartość pierwszego bita się nie zmienia, więc pozostałe macierze poza przekątną są zerami; 
* gdy pierwszy bit jest równy 0, stosowana jest identyczność na drugim bicie, więc górno-lewa macierz to $ I$; a
* gdy pierwszy bit jest równy 1, stosowany jest operator $ M $ na drugim bicie, więc dolno-prawa macierz to $ M $.

### Operator sterowany aktywowany, gdy w stanie 0


Dla danego jednobitowego operatora $ M $, **jak możemy uzyskać następujący operator** używając operatora $ CM $?

$ C_0M = \mymatrix{c|c}{ M & \mathbf{0} \\ \hline \mathbf{0} & I } $

Operatory kontrolowane są zdefiniowane tak, aby zostały wywołane, gdy bit kontrolny znajduje się w stanie 1. W tym przykładzie oczekujemy, że zostaną wywołane, gdy bit kontrolny znajduje się w stanie 0.

Możemy tu skorzystać z prostego triku. Najpierw stosujemy operator NOT do pierwszego bitu, a następnie operator CM, i ponownie operator NOT. W ten sposób gwarantujemy, że $ M $ zostanie zastosowany do drugiego bitu, jeśli pierwszy bit jest w stanie 0 i nie zrobi nic, jeśli pierwszy bit jest w stanie 1. W skrócie:

$$ C_0M = (X \otimes I) \cdot (CM) \cdot ( X \otimes I ). $$

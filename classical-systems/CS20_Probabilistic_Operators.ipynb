Pamiętaj monety Asji i jej protokół rzucania monetą.

$
GameCoins = \begin{array}{c|cc} \hookleftarrow & \mathbf{Head} & \mathbf{Tail} \\ \hline \mathbf{Head} & 0.6 & 0.3\\  \mathbf{Tail} & 0.4 & 0.7  \end{array} \equiv \begin{array}{c|cc} \hookleftarrow & \mathbf{0} & \mathbf{1} \\ \hline \mathbf{0} & 0.6 & 0.3 \\  \mathbf{1} & 0.4 & 0.7  \end{array}
$


Śledzimy wyniki Asji po dwóch rzutach monetą.

<b> Na początku: </b>

<i> Zapamiętaj protokół:
<ol> 
    <li> za każdym razem, gdy wypadnie orzeł, rzuca jednym euro; </li>
    <li> za każdym razem, gdy wypadnie reszka, rzuca jednym centem; oraz </li>
    <li> zaczyna od rzucenia jednym euro, zakładając, że wypadł jej niewidoczny orzeł </li>
</ol>
</i>

Zaczyna w stanie 0: $ v_0 = \myvector{1 \\ 0} $.

Stan 0 reprezentuje Orła, a stan 1 reprezentuje Reszkę.


<b>Po pierwszym rzucie monetą:</b>

$  \mypar{ 
        GameCoins = \begin{array}{c|cc} \hookleftarrow & \mathbf{Head} & \mathbf{Tail} \\ \hline \mathbf{Head} & 0.6 & 0.3\\  \mathbf{Tail} & 0.4 & 0.7  \end{array}
        }
    \mypar{
        CurrentState=\myvector{1 \\ 0}
    } $
    
Każda wartość nowego wektora jest obliczana przez wiersz i aktualny stan:

$
\myvector{  0.6 \cdot 1 \\ + \\ 0.3 \cdot 0 \\ \hline 0.4 \cdot 1 \\ + \\ 0.7 \cdot 0  }
=
\myvector{ 0.6 + 0 \\ 0.4 + 0 }
=
\myvector{0.6 \\ 0.4}.
$


<b>Po drugim rzucie monetą:</b>

$  \mypar{ 
        GameCoins = \begin{array}{c|cc} \hookleftarrow & \mathbf{Orzeł} & \mathbf{Reszka} \\ \hline \mathbf{Orzeł} & 0.6 & 0.3\\  \mathbf{Reszka} & 0.4 & 0.7  \end{array}
        }
    \mypar{
        CurrentState=\myvector{0.6 \\ 0.4}
    } $
    
Każdy element nowego wektora jest obliczany przez wiersz i aktualny stan:

$
\myvector{  0.6 \cdot 0.6 \\ + \\ 0.3 \cdot 0.4 \\ \hline 0.4 \cdot 0.6 \\ + \\ 0.7 \cdot 0.4  }
=
\myvector{ 0.36 + 0.12 \\ 0.24 + 0.28 }
=
\myvector{0.48 \\ 0.52}.
$


<b>Protokół rzucania monetą</b> przez Asję jest <i>operatorem probabilistycznym</i>.

Podobnie jak w przypadku każdego operatora, w zależności od bieżącego stanu, protokół rzucania monetą przez Asję określa następny stan.

$$
    \myvector{1 \\ 0} \xrightarrow{\mbox{Protokół rzucania monetą przez Asję}} \myvector{0.6 \\ 0.4}
    \xrightarrow{\mbox{Protokół rzucania monetą przez Asję}}  \myvector{0.48 \\ 0.52}.
$$


```markdown
<font color="blue"><b>Operator probabilistyczny ewoluuje system z stanu probabilistycznego do stanu probabilistycznego.</b></font>

Protokół rzucania monetą przez Asję przekształca $ \myvector{ 0.8 \\ 0.2 } $ w $ \myvector{ 0.54 \\ 0.46 } $.

Podczas obliczania nowego stanu używamy tabeli $  GameCoins = \begin{array}{c|cc} \hookleftarrow & \mathbf{Head} & \mathbf{Tail} \\ \hline \mathbf{Head} & 0.6 & 0.3\\  \mathbf{Tail} & 0.4 & 0.7  \end{array} $ oraz bieżącego stanu $ \myvector{ 0.8 \\ 0.2 } $:

$$
    \myvector{ \myarray{c}{0.6 \cdot 0.8 \\ + \\ 0.3 \cdot 0.2} \\ \hline \myarray{c}{0.4 \cdot 0.8 \\ + \\ 0.7 \cdot 0.2} } = \myvector{ \myarray{c}{0.48 + 0.06} \\ \myarray{c}{0.32 + 0.14}  } = \myvector{ 0.54 \\ 0.46 }.
$$
```


### Operator probabilistyczny

Operator probabilistyczny może być przedstawiony jako kwadratowa tabela lub macierz.

Wpisy w operatorze probabilistycznym reprezentują prawdopodobieństwa przejścia między stanami.

Dlatego <b>każdy wpis jest nieujemny.</b>

Każda kolumna reprezentuje prawdopodobieństwa przejścia z danego stanu do wszystkich stanów. Dlatego <b>suma wszystkich wpisów w każdej kolumnie wynosi 1</b>, czyli prawdopodobieństwo 1 jest rozłożone na wszystkie stany.

Dowolna macierz spełniająca te dwie właściwości jest nazywana <b>macierzą stochastyczną</b>.

Operator probabilistyczny jest macierzą stochastyczną, i vice versa.

<i> Zauważ, że operator każdego systemu liniowego jest reprezentowany jako macierz. </i>


<h3> Ewolucja probabilistyczna </h3>

Stan probabilistyczny to wektor stochastyczny, powiedzmy $ v $.

Operator probabilistyczny to macierz stochastyczna, powiedzmy $ A $.

Jeśli operator probabilistyczny $ A $ jest zastosowany do stanu probabilistycznego $ v $, nowy stan, powiedzmy $v'$, jest obliczany jako 

$  v' = A \cdot v. $

<i> Zauważ, że ewolucja systemu liniowego jest reprezentowana przez mnożenie macierzy przez wektor. </i>


Jeśli reprezentujemy $ GameCoins $ jako macierz: $ \mymatrix{cc}{ 0.6 & 0.3 \\ 0.4 & 0.7 } $, to nowy stan probabilistyczny jest obliczany jako

$
    \myvector{ 0.54 \\ 0.46 } = \mymatrix{cc}{ 0.6 & 0.3 \\ 0.4 & 0.7 } \myvector{ 0.8 \\ 0.2 }.
$

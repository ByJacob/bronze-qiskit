Załóżmy, że mamy dwie bity probabilistyczne, a nasze stanu probabilistyczne odpowiednio to

$ \myvector{0.2 \\ 0.8} \mbox{ oraz } \myvector{0.6 \\ 0.4 }. $

Jeśli połączymy obie bity jako pojedynczy system, to jaki będzie stan tego połączonego systemu?


W sumie mamy cztery różne stany. Możemy je nazwać następująco:
<ul>
    <li>00: oba bity są w stanie 0</li>
    <li>01: pierwszy bit jest w stanie 0, a drugi bit jest w stanie 1</li>
    <li>10: pierwszy bit jest w stanie 1, a drugi bit jest w stanie 0</li>
    <li>11: oba bity są w stanie 1</li>
</ul>


### Zadanie 1

**Dyskusja i analiza:**

Jakie są prawdopodobieństwa bycia w stanach $00$, $01$, $10$ i $11$?

Jak można przedstawić te prawdopodobieństwa jako wektor kolumnowy?


### Reprezentacja dla stanów 0 i 1

Wektorowa reprezentacja stanu 0 to $ \myvector{1 \\ 0} $. Podobnie, wektorowa reprezentacja stanu 1 to $ \myvector{0 \\ 1} $.

Używamy $ \pstate{0} $ do reprezentacji $ \myvector{1 \\ 0} $ oraz $ \pstate{1} $ do reprezentacji $ \myvector{0 \\ 1} $.

Zatem stan probabilistyczny $ \myvector{0.2 \\ 0.8} $ jest również reprezentowany jako $ 0.2 \pstate{0} + 0.8 \pstate{1} $.

Podobnie, stan probabilistyczny $ \myvector{0.6 \\ 0.4} $ jest również reprezentowany jako $ 0.6 \pstate{0} + 0.4 \pstate{1} $.


### Systemy złożone

Gdy dwa systemy są skomponowane, ich stany są tensorem obliczone, aby obliczyć stan systemu złożonego.

Prawdopodobieństwem stanu pierwszego bitu jest $ \myvector{0.2 \\ 0.8} = 0.2 \pstate{0} + 0.8 \pstate{1} $.

Prawdopodobieństwem stanu drugiego bitu jest $ \myvector{0.6 \\ 0.4} = 0.6 \pstate{0} + 0.4 \pstate{1} $.

Wówczas stan prawdopodobieństwa systemu złożonego to $ \big( 0.2 \pstate{0} + 0.8 \pstate{1} \big) \otimes \big( 0.6 \pstate{0} + 0.4 \pstate{1} \big) $.

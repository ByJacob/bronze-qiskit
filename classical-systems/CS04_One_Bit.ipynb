$ 0,~1,~2,~3,~4,~5,~6,~7,~8, \mbox{ i } 9 $.

W informatyce, z drugiej strony, powszechnie używanym systemem jest system dwójkowy, który ma tylko dwie cyfry:

$ 0 $ i $ 1 $.

Bit (lub cyfra binarna) to podstawowa jednostka informacji używana w informatyce i teorii informacji.

Można go również traktować jako najmniejszą "użyteczną" jednostkę pamięci, która ma dwie stany o nazwach 0 i 1.

W dowolnym momencie bit może znajdować się w stanie 0 lub w stanie 1.


<h3> Czterech operatorów </h3>

Ile różnych operatorów można zdefiniować na pojedynczym bicie?

<i>Operator, w zależności od aktualnego stanu bitu, aktualizuje stan bitu (wynik może być taki sam).</i>

Możemy zastosować cztery różne operatory do pojedynczego bitu:
<ol>
    <li> Tożsamość: $ I(0) = 0 $ i $ I(1) = 1 $ </li>
    <li> Negacja: $ NOT(0) = 1 $ i $ NOT(1) = 0 $ </li>
    <li> Stała (Zero): $ ZERO(0) = 0 $ i $ ZERO(1) = 0 $ </li>
    <li> Stała (Jeden): $ ONE(0) = 1 $ i $ ONE(1) = 1 $ </li>
</ol>
Pierwszy operator nazywany jest TOŻSAMOŚĆ, ponieważ nie zmienia zawartości/wartości bitu.

Drugi operator nazywa się NEGACJA, ponieważ odwraca (odwraca) wartość bitu.

<i>Zauważ, że 0 i 1 odnoszą się również do wartości logicznych False i True, odpowiednio, gdzie False jest negacją True, a True jest negacją False.</i>

Trzeci (odpowiednio, czwarty) operator zwraca stałą wartość 0 (odpowiednio, 1), bez względu na wejście.


### Reprezentacja tabelaryczna

W naszych reprezentacjach, kierunek przejść jest od góry do lewej, tj. stany początkowe znajdują się u góry, a stany końcowe po lewej:

$
    \begin{array}{c|c}
        & stany~początkowe \\ \hline
        stany~końcowe&\hookleftarrow
    \end{array}
$

Możemy reprezentować przejścia każdego operatora za pomocą tabeli:

$
I = \begin{array}{lc|cc} 
     & & początkowe & stany \\ 
    & \hookleftarrow & \mathbf{0} & \mathbf{1} \\ \hline 
    końcowe & \mathbf{0} & \mbox{idzie~do} & \emptyset \\  
    stany & \mathbf{1} & \emptyset & \mbox{idzie~do}  \end{array} ,
$
gdzie 
- nagłówek (pierwszy wiersz) reprezentuje wartości początkowe, a
- pierwsza kolumna reprezentuje wartości końcowe.



Możemy również zdefiniować przejścia numerycznie:
- używamy 1, jeśli istnieje przejście między dwoma stanami, oraz,
- używamy 0, jeśli nie istnieje przejście między dwoma stanami.

$
I = \begin{array}{lc|cc} 
    & & początkowe & stany \\ 
    & \hookleftarrow & \mathbf{0} & \mathbf{1} \\ \hline 
    końcowe & \mathbf{0} & 1 & 0 \\  
    stany & \mathbf{1} & 0 & 1  \end{array}
$


Wartości w <b>pogrubieniu</b> oznaczają początkowe i końcowe wartości bitów. Wartości niepogrubione reprezentują przejścia.
<ul>
    <li> Niepogrubione 1 w lewym górnym rogu reprezentuje przejście $ 0 \rightarrow 0 $. </li>
    <li> Niepogrubione 1 w prawym dolnym rogu reprezentuje przejście $ 1 \rightarrow 1 $. </li> 
    <li> Niepogrubione 0 w prawym górnym rogu oznacza brak przejścia z 1 do 0. </li>
    <li> Niepogrubione 0 w lewym dolnym rogu oznacza brak przejścia z 0 do 1. </li>
</ul>
Czytelnik może myśleć, że wartości 0 i 1 reprezentują przejścia jako Fałsz (Wyłączone) i Prawda (Włączone), odpowiednio.

Podobnie, możemy reprezentować inne operatory jak poniżej:

$
NOT = \begin{array}{lc|cc} & & początkowe & stany \\ & \hookleftarrow & \mathbf{0} & \mathbf{1} \\ \hline final & \mathbf{0} & 0 & 1 \\  
    stany & \mathbf{1} & 1 & 0  \end{array}
~~~
ZERO = \begin{array}{lc|cc} & & początkowe & stany \\ & \hookleftarrow & \mathbf{0} & \mathbf{1} \\ \hline final & \mathbf{0} & 1 & 1 \\  
    stany & \mathbf{1} & 0 & 0  \end{array}
~~~
ONE = \begin{array}{lc|cc} & & początkowe & stany \\ & \hookleftarrow & \mathbf{0} & \mathbf{1} \\ \hline final & \mathbf{0} & 0 & 0 \\  
    stany & \mathbf{1} & 1 & 1  \end{array}
.
$

# Reversybilność i Nieodwracalność

Po zastosowaniu operatora Identity lub NOT, łatwo możemy określić początkową wartość, sprawdzając końcową wartość.
- W przypadku operatora Identity po prostu mówimy tę samą wartość.
- W przypadku operatora NOT po prostu mówimy inną wartość, tj. jeśli końcowa wartość to 0 (lub 1), to mówimy 1 (lub 0).

Jednak nie możemy poznać początkowej wartości, sprawdzając końcową wartość po zastosowaniu operatora ZERO lub ONE.

Na podstawie tej obserwacji możemy sklasyfikować operatory na dwa rodzaje: <i>Odwrotnie</i> i <i>Nieodwracalne</i>.
- Jeśli możemy odzyskać początkową wartość(y) z wartości(y) końcowej(ej), to operator jest nazywany odwracalnym, tak jak operatory Identity i NOT.
- Jeśli nie możemy poznać początkowej wartości(y) z wartości(y) końcowej(ej), to operator jest nazywany nieodwracalnym, tak jak operatory ZERO i ONE.

**Ta klasyfikacja jest ważna, ponieważ operatory ewolucji kwantowej są odwracalne (o ile system jest zamknięty).**

Operatory Identity i NOT są dwoma fundamentalnymi operatorami kwantowymi.
